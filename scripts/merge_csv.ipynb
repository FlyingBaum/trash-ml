```python
import os
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame()

#Trash + Tourist data
df_trash_tourists = pd.read_csv('../data/trash-amounts-tourist.csv')
df_trash_tourists.drop(columns=["Avg_Daily_Arrivals","Avg_Daily_Stays", "Tour_Arrivals"], inplace=True)
df = pd.concat([df, df_trash_tourists], axis=1)

# Set date as index
df = df.set_index('Date')
df.head()
```

```python
# weather
df_weather = pd.read_csv('../data/weather.csv')
df_weather.drop(columns=["snow(mm)"], inplace=True)
df_weather = df_weather.set_index('Date')

df = df.join(df_weather)

df.head()
```

```python
df.columns
```

```python
# yards
yards = df["Yard"].unique()
for yard in yards:
    df[yard] = np.where(df["Yard"] == yard, True, False)

```

```python
#trash type
types = df["Type"].unique()
for t in types:
    df[t] = np.where(df["Type"] == t, True, False)
```

```python
# tour numbers
tour_numbers = df["Tour"].unique()
for tn in tour_numbers:
    df["Tour_"+str(tn)] = np.where(df["Tour"] == tn, True, False)
```

```python
#holiday and weekdays
df_holiday_weekday = pd.read_csv('../data/holidays-and-weekdays.csv')
df_holiday_weekday.rename(columns={'Datum': 'Date'}, inplace=True)
df_holiday_weekday['Date'] = pd.to_datetime(df_holiday_weekday['Date'])
df_holiday_weekday['Date'] = df_holiday_weekday['Date'].dt.strftime('%d.%m.%y')
df_holiday_weekday.set_index('Date', inplace=True)
df = df.join(df_holiday_weekday)
# df = pd.merge(df, df_holiday_weekday, left_on=['Date'], how='left')
# df = pd.concat([df, df_holiday_weekday], ignore_index=True)
df_holiday_weekday.head(20)
```
