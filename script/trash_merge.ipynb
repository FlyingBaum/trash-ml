```python
import pandas as pd
```

```python
trash_dtypes = {
    'Month': str,
    'CW': str,
    'Year': str,
    'Date': str,
    'Yard': str,
    'Shift': str,
    'Tour': str,
    'Tonnage': str,
    'Type': str
}

df1 = pd.read_csv('../data/trash-amounts.csv', delimiter=';', skiprows=1, names=['Month', 'CW', 'Year', 'Date', 'Yard', 'Shift', 'Tour', 'Tonnage', 'Type'], dtype=trash_dtypes)

df1.head(20)
```

```python
df2 = pd.read_csv('../data/tourism_clean.csv', dtype={'Month': str})

df2.head()
```

```python
df1['Month'] = df1['Month'].astype(str)
df1['Year'] = df1['Year'].astype(str)
df2['Month'] = df2['Month'].astype(str)
df2['Year'] = df2['Year'].astype(str)

# Perform the merge
df_merged = pd.merge(df1, df2, on=['Month', 'Year'], how='left')
df_merged = df_merged[['Month', 'CW', 'Year', 'Date', 'Yard', 'Shift', 'Tour', 'Tonnage', 'Type', 'Avg_Daily_Arrivals', 'Avg_Daily_Stays']]
df_merged.head()
```

```python
df_merged.to_csv('../data/trash-amounts-tourist.csv', index=False)
```
